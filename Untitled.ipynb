```python
import pandas as pd
import folium
from folium.plugins import BeautifyIcon
import math

# Load data
df = pd.read_csv("C:/Users/RAKHA ALCANDER/OneDrive/Documents/MAGANG PLN/Pak Arung/PROJEK/Desa Kajarharjo Banyuwangi.csv", sep=';')
df[['Latitude', 'Longitude', 'Label']] = df['Latitude'].str.split(',', expand=True)
df['Latitude'] = df['Latitude'].str.strip().astype(float)
df['Longitude'] = df['Longitude'].str.strip().astype(float)
df['Label'] = df['Label'].str.strip()

def calculate_bearing(lat1, lon1, lat2, lon2):
    """Calculate bearing between two points"""
    lat1, lat2 = math.radians(lat1), math.radians(lat2)
    delta_lon = math.radians(lon2 - lon1)
    
    x = math.sin(delta_lon) * math.cos(lat2)
    y = math.cos(lat1) * math.sin(lat2) - math.sin(lat1) * math.cos(lat2) * math.cos(delta_lon)
    
    bearing = math.degrees(math.atan2(x, y))
    return (bearing + 360) % 360

def angle_diff(b1, b2):
    """Calculate the smaller angle difference between two bearings"""
    diff = abs(b1 - b2)
    return min(diff, 360 - diff)

def classify_angle(angle):
    """Classify angle into TM categories - COMPLETE VERSION"""
    if pd.isna(angle) or angle is None:
        return None
    
    angle = float(angle)
    
    # Debug print untuk melihat nilai angle
    print(f"Classifying angle: {angle}")
    
    # Klasifikasi LENGKAP berdasarkan sudut belok (0-180 derajat)
    if 0 <= angle <= 15:
        return "TM1"    # Hampir lurus
    elif 16 <= angle <= 30:
        return "TM2"    # Belok sedikit
    elif 31 <= angle <= 60:
        return "TM5"    # Belok sedang
    elif 61 <= angle <= 180:
        return "TM10"   # Belok tajam hingga U-turn
    else:
        # Seharusnya tidak ada sudut di luar 0-180 derajat
        print(f"WARNING: Unexpected angle value: {angle}")
        return "TM10"   # Default ke TM10 untuk safety

# Initialize column
df["Sudut (derajat)"] = None

# Calculate angles - improved version
for i in range(1, len(df)-1):
    try:
        # Get coordinates
        lat1, lon1 = df.loc[i-1, ["Latitude", "Longitude"]]
        lat2, lon2 = df.loc[i, ["Latitude", "Longitude"]]
        lat3, lon3 = df.loc[i+1, ["Latitude", "Longitude"]]
        
        # Check if coordinates are valid
        if pd.isna(lat1) or pd.isna(lon1) or pd.isna(lat2) or pd.isna(lon2) or pd.isna(lat3) or pd.isna(lon3):
            print(f"Invalid coordinates at index {i}")
            continue
            
        # Calculate bearings
        bearing1 = calculate_bearing(lat1, lon1, lat2, lon2)
        bearing2 = calculate_bearing(lat2, lon2, lat3, lon3)
        
        # Calculate angle difference
        angle = angle_diff(bearing1, bearing2)
        
        # Debug information
        print(f"Point {i}: bearing1={bearing1:.2f}, bearing2={bearing2:.2f}, angle={angle:.2f}")
        
        df.at[i, "Sudut (derajat)"] = round(angle, 2)
        
    except Exception as e:
        print(f"Error calculating angle for point {i}: {e}")
        continue

# Apply classification
df["Kategori"] = df["Sudut (derajat)"].apply(classify_angle)

# Set last point category
df.at[df.index[-1], "Kategori"] = "TM4"

# Debug: Print results
print("\nResults:")
print(df[["Label", "Sudut (derajat)", "Kategori"]].to_string())

# Check for unclassified points
unclassified = df[df["Kategori"].isna()]
if not unclassified.empty:
    print(f"\nUnclassified points: {len(unclassified)}")
    print(unclassified[["Label", "Sudut (derajat)", "Kategori"]])

# Count by category
print(f"\nCategory counts:")
print(df["Kategori"].value_counts(dropna=False))

# Visualisasi dengan Folium
color_map = {
    "TM1": "blue",
    "TM2": "orange", 
    "TM5": "purple",
    "TM10": "red",
    "TM4": "green",
    None: "gray"
}

m = folium.Map(location=[df["Latitude"].mean(), df["Longitude"].mean()], zoom_start=17)

for _, row in df.iterrows():
    # Handle None values in popup
    sudut_text = row['Sudut (derajat)'] if pd.notna(row['Sudut (derajat)']) else "N/A"
    kategori_text = row['Kategori'] if pd.notna(row['Kategori']) else "N/A"
    
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=f"{row['Label']}<br>Sudut: {sudut_text}<br>Kategori: {kategori_text}",
        icon=BeautifyIcon(
            icon_shape='marker',
            border_color=color_map.get(row["Kategori"], "gray"),
            text_color=color_map.get(row["Kategori"], "gray"),
            background_color="white",
            number=row["Label"]
        )
    ).add_to(m)

# Add polyline
folium.PolyLine(df[["Latitude", "Longitude"]].values.tolist(), color="black", weight=2.5).add_to(m)

# Display map
m
```
